```python
# 27. Use Recursion to calculate the factorial of a number.

def factorial(n):
  if n == 0 or n == 1:
    return 1
  else:
    return n * factorial(n - 1)


# taking the input from the user
num = int(input("Enter a number:"))

if num < 0:
  print("Factorial is not defined for negative numbers.")
else:
  print(f"The factorial of {num} is {factorial(num)}")
```

```python
# 52. WAP to perform push & pop operation in Stack.

stack =[]

# function to push Elements
def push():
   val = int(input("Enter Element to Push: "))
   stack.append(val)
   print("Element Pushed Successfully")

# function to pop elemnt
def pop():
  if len(stack) == 0:
      print("Stack is Empty")

  else:
    removed = stack.pop()
    print("Removed Element:", removed)


while True:

    print("\n1. Push")
    print("2. Pop")
    print("3. Display")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        push()

    elif choice == 2:
        pop()

    elif choice == 3:
        print("Stack:", stack)

    elif choice == 4:
        print("Exiting...")
        break

    else:
        print("Invalid choice")
```

```python
# 53. WAP program to add & remove in a Queue .

queue = []

# Function to add elements to the queue (enqueue)
def enqueue():
    val = input("Enter element to add to the queue: ")
    queue.append(val)
    print(f"'{val}' added to the queue successfully.")

# Function to remove elements from the queue (dequeue)
def dequeue():
    if not queue:
        print("Queue is empty. Cannot remove elements.")
    else:
        removed_val = queue.pop(0) # FIFO: remove from the front
        print(f"'{removed_val}' removed from the queue.")

# Main loop for user interaction
while True:
    print("\n--- Queue Operations ---")
    print("1. Enqueue (Add element)")
    print("2. Dequeue (Remove element)")
    print("3. Display Queue")
    print("4. Exit")

    choice = input("Enter your choice (1/2/3/4): ")

    if choice == '1':
        enqueue()
    elif choice == '2':
        dequeue()
    elif choice == '3':
        print("Current Queue:", queue)
    elif choice == '4':
        print("Exiting Queue program.")
        break
    else:
        print("Invalid choice. Please enter 1, 2, 3, or 4.")
```
